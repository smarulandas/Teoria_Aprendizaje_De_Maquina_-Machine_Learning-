1) Simulación para 2 cajas, una caja $roja(r)$ y una $azul(b)$. <br><br>
$\hspace{1cm}$ - En la caja roja hay 2 apple(a) y 6 orange(o).
<br>

$\hspace{1cm}$ - En la caja azul hay 1 orange(o) y 3 apple(a).
<br><br>

- Con una probabiliad Prior de $P(B=r) = 0.4$ y $P(B=b) = 0.6$. <br>
-Y sabiendo que $P(B=r) + P(B=b) = 1.$
<br><br>

Hallar las siguientes probabilidades:
<br>
$\hspace{1cm}$ - Verosimilitud <br>
$\hspace{1cm}$ - Prior <br>
$\hspace{1cm}$ - Evidencia <br>
$\hspace{1cm}$ - Posterior <br>

Es decir, realizar una silumación con 2 cajas y y 2 frutas para hallar la probabilidad Posterior.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los datos del problema
# Cajas: r (roja), b (azul)
# Frutas: a (apple), o (orange)

# Probabilidades prior
p_b_r = 0.4  # P(B=r), probabilidad de elegir la caja roja
p_b_b = 0.6  # P(B=b), probabilidad de elegir la caja azul

# Contenido de las cajas
# Caja roja: 2 manzanas (a) y 6 naranjas (o)
# Caja azul: 3 manzanas (a) y 1 naranja (o)
red_box = {'a': 2, 'o': 6}
blue_box = {'a': 3, 'o': 1}

# Probabilidades totales en cada caja
total_red = sum(red_box.values())
total_blue = sum(blue_box.values())

# 1. Cálculo de la Verosimilitud (Likelihood)
# P(F|B) - probabilidad de sacar una fruta dada una caja

# P(F=a|B=r) - probabilidad de sacar una manzana de la caja roja
p_a_given_r = red_box['a'] / total_red

# P(F=o|B=r) - probabilidad de sacar una naranja de la caja roja
p_o_given_r = red_box['o'] / total_red

# P(F=a|B=b) - probabilidad de sacar una manzana de la caja azul
p_a_given_b = blue_box['a'] / total_blue

# P(F=o|B=b) - probabilidad de sacar una naranja de la caja azul
p_o_given_b = blue_box['o'] / total_blue

# 2. Probabilidad Prior
# Ya definida: P(B=r)=0.4, P(B=b)=0.6

# 3. Cálculo de la Evidencia
# P(F) - probabilidad total de obtener cierta fruta
# P(F=a) = P(F=a|B=r)*P(B=r) + P(F=a|B=b)*P(B=b)
p_a = p_a_given_r * p_b_r + p_a_given_b * p_b_b

# P(F=o) = P(F=o|B=r)*P(B=r) + P(F=o|B=b)*P(B=b)
p_o = p_o_given_r * p_b_r + p_o_given_b * p_b_b

# 4. Cálculo de Posterior usando el Teorema de Bayes
# P(B|F) = P(F|B)*P(B) / P(F)

# P(B=r|F=a) - probabilidad de haber elegido la caja roja dado que sacamos una manzana
p_r_given_a = (p_a_given_r * p_b_r) / p_a

# P(B=b|F=a) - probabilidad de haber elegido la caja azul dado que sacamos una manzana
p_b_given_a = (p_a_given_b * p_b_b) / p_a

# P(B=r|F=o) - probabilidad de haber elegido la caja roja dado que sacamos una naranja
p_r_given_o = (p_o_given_r * p_b_r) / p_o

# P(B=b|F=o) - probabilidad de haber elegido la caja azul dado que sacamos una naranja
p_b_given_o = (p_o_given_b * p_b_b) / p_o

# Mostrar resultados
print("Resultados del cálculo de probabilidades:")
print("\n1. Verosimilitud (Likelihood):")
print(f"P(F=a|B=r) = {p_a_given_r:.4f}")
print(f"P(F=o|B=r) = {p_o_given_r:.4f}")
print(f"P(F=a|B=b) = {p_a_given_b:.4f}")
print(f"P(F=o|B=b) = {p_o_given_b:.4f}")

print("\n2. Prior:")
print(f"P(B=r) = {p_b_r:.4f}")
print(f"P(B=b) = {p_b_b:.4f}")

print("\n3. Evidencia:")
print(f"P(F=a) = {p_a:.4f}")
print(f"P(F=o) = {p_o:.4f}")

print("\n4. Posterior:")
print(f"P(B=r|F=a) = {p_r_given_a:.4f}")
print(f"P(B=b|F=a) = {p_b_given_a:.4f}")
print(f"P(B=r|F=o) = {p_r_given_o:.4f}")
print(f"P(B=b|F=o) = {p_b_given_o:.4f}")

# Verificar que las probabilidades suman 1
print("\nVerificaciones:")
print(f"P(B=r) + P(B=b) = {p_b_r + p_b_b:.4f}")
print(f"P(F=a) + P(F=o) = {p_a + p_o:.4f}")
print(f"P(B=r|F=a) + P(B=b|F=a) = {p_r_given_a + p_b_given_a:.4f}")
print(f"P(B=r|F=o) + P(B=b|F=o) = {p_r_given_o + p_b_given_o:.4f}")

# Visualización de resultados
def plot_probability_bars():
    labels = ['P(F=a|B=r)', 'P(F=o|B=r)', 'P(F=a|B=b)', 'P(F=o|B=b)']
    likelihood_vals = [p_a_given_r, p_o_given_r, p_a_given_b, p_o_given_b]

    plt.figure(figsize=(12, 8))

    # Gráfico de Verosimilitud
    plt.subplot(2, 2, 1)
    plt.bar(labels, likelihood_vals, color=['red', 'red', 'blue', 'blue'])
    plt.title('Verosimilitud (Likelihood)')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)

    # Gráfico de Prior
    plt.subplot(2, 2, 2)
    plt.bar(['P(B=r)', 'P(B=b)'], [p_b_r, p_b_b], color=['red', 'blue'])
    plt.title('Prior')
    plt.ylim(0, 1)

    # Gráfico de Evidencia
    plt.subplot(2, 2, 3)
    plt.bar(['P(F=a)', 'P(F=o)'], [p_a, p_o], color=['green', 'orange'])
    plt.title('Evidencia')
    plt.ylim(0, 1)

    # Gráfico de Posterior
    plt.subplot(2, 2, 4)
    plt.bar(['P(B=r|F=a)', 'P(B=b|F=a)', 'P(B=r|F=o)', 'P(B=b|F=o)'],
            [p_r_given_a, p_b_given_a, p_r_given_o, p_b_given_o],
            color=['red', 'blue', 'red', 'blue'])
    plt.title('Posterior')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

# Descomentar para mostrar los gráficos
#plot_probability_bars()
```

2) Simulación para n cajas y m frutas

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate

class BayesianBoxModel:
    def __init__(self):
        """Inicialización del modelo Bayesiano para n cajas y m frutas."""
        self.boxes = {}  # Diccionario para almacenar el contenido de cada caja
        self.priors = {}  # Probabilidades prior para cada caja
        self.fruits = set()  # Conjunto de todos los tipos de frutas

    def add_box(self, box_id, fruit_counts, prior_prob=None):
        """
        Añade una caja al modelo.

        Args:
            box_id: Identificador único para la caja
            fruit_counts: Diccionario con {fruta: cantidad} para esta caja
            prior_prob: Probabilidad prior de esta caja (opcional)
        """
        self.boxes[box_id] = fruit_counts

        # Añadir todos los tipos de fruta al conjunto de frutas
        for fruit in fruit_counts:
            self.fruits.add(fruit)

        # Si se proporciona un prior, guardarlo
        if prior_prob is not None:
            self.priors[box_id] = prior_prob

    def set_priors(self, priors_dict=None):
        """
        Establece las probabilidades prior para cada caja.
        Si no se proporcionan, se asume una distribución uniforme.

        Args:
            priors_dict: Diccionario con {box_id: prior_prob}
        """
        if priors_dict is None:
            # Distribución uniforme si no se proporcionan priors
            n_boxes = len(self.boxes)
            for box_id in self.boxes:
                self.priors[box_id] = 1.0 / n_boxes
        else:
            # Usar los priors proporcionados
            self.priors = priors_dict

        # Normalizar los priors para asegurarse de que suman 1
        total_prior = sum(self.priors.values())
        if abs(total_prior - 1.0) > 1e-10:  # Permitir un pequeño error numérico
            for box_id in self.priors:
                self.priors[box_id] /= total_prior

    def calculate_likelihoods(self):
        """
        Calcula las probabilidades de verosimilitud P(F=f|B=b) para cada combinación
        de fruta f y caja b.

        Returns:
            dict: Diccionario de verosimilitudes {box_id: {fruit: likelihood}}
        """
        likelihoods = {}

        for box_id, fruit_counts in self.boxes.items():
            box_total = sum(fruit_counts.values())
            likelihoods[box_id] = {}

            for fruit in self.fruits:
                if fruit in fruit_counts:
                    likelihoods[box_id][fruit] = fruit_counts[fruit] / box_total
                else:
                    likelihoods[box_id][fruit] = 0.0

        return likelihoods

    def calculate_evidence(self, likelihoods):
        """
        Calcula la probabilidad de evidencia P(F=f) para cada fruta f.

        Args:
            likelihoods: Diccionario de verosimilitudes calculado previamente

        Returns:
            dict: Diccionario de evidencias {fruit: evidence}
        """
        evidence = {fruit: 0.0 for fruit in self.fruits}

        for fruit in self.fruits:
            for box_id in self.boxes:
                evidence[fruit] += likelihoods[box_id][fruit] * self.priors[box_id]

        return evidence

    def calculate_posteriors(self, likelihoods, evidence):
        """
        Calcula las probabilidades posteriores P(B=b|F=f) para cada combinación
        de caja b y fruta f.

        Args:
            likelihoods: Diccionario de verosimilitudes calculado previamente
            evidence: Diccionario de evidencias calculado previamente

        Returns:
            dict: Diccionario de posteriores {fruit: {box_id: posterior}}
        """
        posteriors = {fruit: {} for fruit in self.fruits}

        for fruit in self.fruits:
            for box_id in self.boxes:
                if evidence[fruit] > 0:
                    posteriors[fruit][box_id] = (likelihoods[box_id][fruit] *
                                                self.priors[box_id]) / evidence[fruit]
                else:
                    posteriors[fruit][box_id] = 0.0

        return posteriors

    def calculate_all_probabilities(self):
        """
        Calcula todas las probabilidades (verosimilitud, prior, evidencia, posterior)
        y las devuelve en un diccionario.

        Returns:
            dict: Diccionario con todas las probabilidades
        """
        likelihoods = self.calculate_likelihoods()
        evidence = self.calculate_evidence(likelihoods)
        posteriors = self.calculate_posteriors(likelihoods, evidence)

        return {
            'likelihoods': likelihoods,
            'priors': self.priors,
            'evidence': evidence,
            'posteriors': posteriors
        }

    def display_probabilities(self, probabilities=None):
        """
        Muestra todas las probabilidades en formato de tabla.

        Args:
            probabilities: Diccionario con todas las probabilidades (opcional)
        """
        if probabilities is None:
            probabilities = self.calculate_all_probabilities()

        # Mostrar verosimilitudes
        print("\n1. Verosimilitud (Likelihood) P(F|B):")
        likelihood_table = []
        header = ["Box/Fruit"] + list(self.fruits)

        for box_id in self.boxes:
            row = [box_id]
            for fruit in self.fruits:
                row.append(f"{probabilities['likelihoods'][box_id][fruit]:.4f}")
            likelihood_table.append(row)

        print(tabulate(likelihood_table, headers=header, tablefmt="grid"))

        # Mostrar priors
        print("\n2. Prior P(B):")
        prior_table = []
        for box_id, prior in probabilities['priors'].items():
            prior_table.append([box_id, f"{prior:.4f}"])

        print(tabulate(prior_table, headers=["Box", "Prior"], tablefmt="grid"))

        # Mostrar evidencias
        print("\n3. Evidencia P(F):")
        evidence_table = []
        for fruit, evidence in probabilities['evidence'].items():
            evidence_table.append([fruit, f"{evidence:.4f}"])

        print(tabulate(evidence_table, headers=["Fruit", "Evidence"], tablefmt="grid"))

        # Mostrar posteriores
        print("\n4. Posterior P(B|F):")
        posterior_table = []
        header = ["Fruit/Box"] + list(self.boxes.keys())

        for fruit in self.fruits:
            row = [fruit]
            for box_id in self.boxes:
                row.append(f"{probabilities['posteriors'][fruit][box_id]:.4f}")
            posterior_table.append(row)

        print(tabulate(posterior_table, headers=header, tablefmt="grid"))

        # Verificar que las sumas sean correctas
        print("\nVerificaciones:")
        print(f"Suma de Priors: {sum(probabilities['priors'].values()):.4f}")
        print(f"Suma de Evidencias: {sum(probabilities['evidence'].values()):.4f}")

        for fruit in self.fruits:
            posterior_sum = sum(probabilities['posteriors'][fruit].values())
            print(f"Suma de Posteriores para {fruit}: {posterior_sum:.4f}")

    def plot_probabilities(self, probabilities=None):
        """
        Visualiza las probabilidades calculadas.

        Args:
            probabilities: Diccionario con todas las probabilidades (opcional)
        """
        if probabilities is None:
            probabilities = self.calculate_all_probabilities()

def crear_modelo_personalizado():
    modelo = BayesianBoxModel()

    # Pedir información sobre las cajas
    try:
        n_boxes = int(input("Ingrese el número de cajas: "))

        for i in range(n_boxes):
            box_id = input(f"Ingrese el identificador para la caja {i+1}: ")

            fruit_counts = {}
            n_fruits = int(input(f"¿Cuántos tipos diferentes de frutas hay en la caja {box_id}? "))

            for j in range(n_fruits):
                fruit_name = input(f"Nombre de la fruta {j+1}: ")
                count = int(input(f"Cantidad de {fruit_name} en la caja {box_id}: "))
                fruit_counts[fruit_name] = count

            modelo.add_box(box_id, fruit_counts)

        # Pedir información sobre los priors
        use_custom_priors = input("¿Desea establecer probabilidades prior personalizadas? (s/n): ").lower() == 's'

        if use_custom_priors:
            valid_priors = False
            while not valid_priors:
                priors = {}
                print("\nLa suma de todas las probabilidades prior debe ser exactamente 1.")
                print("Cada probabilidad debe estar entre 0 y 1.\n")

                try:
                    for box_id in modelo.boxes:
                        while True:
                            prior = float(input(f"Probabilidad prior para la caja {box_id}: "))
                            if 0 <= prior <= 1:
                                priors[box_id] = prior
                                break
                            else:
                                print(f"Error: El prior debe estar entre 0 y 1. Intente nuevamente.")

                    # Verificar que la suma sea 1
                    total_prior = sum(priors.values())
                    if abs(total_prior - 1.0) <= 1e-10:  # Permitir un pequeño error numérico
                        valid_priors = True
                    else:
                        print(f"\nError: La suma de los priors es {total_prior}, pero debe ser exactamente 1.")
                        print("Por favor, ingrese todos los priors nuevamente.\n")
                except ValueError:
                    print("Error: Debe ingresar un número válido. Intente nuevamente.")

            modelo.set_priors(priors)
        else:
            modelo.set_priors()  # Usar distribución uniforme

        # Calcular y mostrar todas las probabilidades
        probabilities = modelo.calculate_all_probabilities()
        modelo.display_probabilities(probabilities)

        # Preguntar si desea visualizar las probabilidades
        if input("¿Desea visualizar las probabilidades? (s/n): ").lower() == 's':
            modelo.plot_probabilities(probabilities)

    except ValueError as e:
        print(f"Error: Asegúrese de ingresar valores numéricos válidos donde corresponda.")
        print(f"Detalles del error: {e}")
        print("Intente ejecutar el programa nuevamente.")

# Para usar esta función, solo necesitas descomentar la siguiente línea:
crear_modelo_personalizado()
```
