## Ejercicio 1: Manipulación y Regex de Diálogos
**Contexto:** Tienes la siguiente cadena de texto que representa un log de un sistema de atención:
`log = "User: [JuanP] | Time: 14:30:05 | Msg: Necesito ayuda con mi suscripción #12345"`

**Instrucción:** Escribe un código en Python que utilice un **único patrón de Expresiones Regulares con grupos con nombre (Named Groups)** para extraer el nombre del usuario, la hora y el número de ticket (el valor numérico después del `#`). Almacena el resultado en un diccionario.

**Pregunta para el examen:** > ¿Qué sucede con tu expresión regular si el nombre del usuario contiene caracteres especiales (ej: `[Juan_P.23]`) y cómo la ajustarías para que sea robusta sin usar grupos genéricos como `.*?`?

````python
#Para solucionar este ejercicio se propone:

# 1) Tomar el string(str) log.
# 2) Buscar dentro de el ciertas partes.
# 3) Extraerlas con regex (r).
# 4) Convertir el resultado a un diccionario.

import re   # Libreria para trabajar con expresiones regulares (buscar patrones dentro de texto, validar formatos, extraer partes especificas)

log = "User: [Juan_P.23] | Time: 14:30:05 | Msg: Necesito ayuda con mi suscripción #12345"

patron = re.compile(      #Se construye el patrón Regex
    r"User:\s*\[(?P<usuario>[^\]]+)\]\s*\|\s*"
    # Se busca texto User
    # \s ->Acepta espacio en blanco y lo consume y * -> cero o mas espacios
    # \[ -> Busca literalmente [, ya que [ significa, abre un conjunto y dice, acepta uno de los siguientes caracteres
    # ?P<usuario> -> llama el nuevo nombre del grupo y [^\]]+ es lo que el grupo va a capturar menos ]
    # \] -> Busca el caracter ]
    #\s*\|\s* -> busca el simbolo | sin importar cuantos espacios esten, para decir que encontro bien el patrón debe hacber un |

    r"Time:\s*(?P<hora>\d{2}:\d{2}:\d{2})\s*\|\s*"

    r"Msg:\s*[^#]*#(?P<ticket>\d+)"
    # [^#]* -> Toma caracteres hasta antes de # y que no sean #.
    # # -> Busca el simbolo #
    # (?P<ticket>\d+) -> grupo con nombre ticket y \d+ toma uno o mas digitos.

)

match = patron.search(log) #busca en log que coincida el patrón completo de todas las r, si alguna no cumple sale NONE(ninguna). (si en time hace falta los segundos sale NONE)

resultado = match.groupdict() if match else {} #si match existe, groupdict convierte los grupos con nombre en diccionario.        Si match es NONE, devuelve {}

print(resultado) # Imprime





#RESPUESTA PREGUNTA PARCIAL

#Si el nombre del usuario contiene caracteres especiales como Juan_P.23, una expresión restrictiva como \w+(Si acepta letras, numeros y guionn bajo, pero no acepta .) podría no capturarlo completo.

# .*? -> toma cualquier cosa, pero detente lo antes posible, esto es menos controlado.

#Para hacerla robusta sin usar .*, se usa un patrón como [^\]]+, que captura cualquier carácter hasta encontrar el corchete de cierre ].
````
